```python
currentPath = os.getcwd()
q = currentPath.split('/')
qq = q[:-3]
newPath = os.path.join(*qq)
newPath
```

```python
# set up environment 
os.environ["DATASET_DIRECTORY"] = "/"+newPath+"/datasets/"+whichDataset
dataYamlPath = os.environ["DATASET_DIRECTORY"]+"/data.yaml"
```

```python
# Modify the data.yaml file's paths:
# Read and Write YAML file
import yaml
import io
with open(dataYamlPath, 'r') as stream:
    data_loaded = yaml.safe_load(stream)
data_loaded['train'] = os.environ["DATASET_DIRECTORY"]+"/"+"train"
data_loaded['val'] = os.environ["DATASET_DIRECTORY"]+"/"+"valid"
with io.open(dataYamlPath, 'w', encoding='utf8') as outfile:
    yaml.dump(data_loaded, outfile, default_flow_style=False, allow_unicode=True)
```
